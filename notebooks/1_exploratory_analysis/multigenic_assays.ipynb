# Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path = "../../data/"
```

# Data

```python
df_assays = pd.read_excel(data_path + "multigenic_assays.xlsx")
df_assays = df_assays.drop(columns=df_assays.columns[1])


target_class = pd.read_csv(data_path + 'target_class.csv')
```

```python
data = pd.merge(df_assays, target_class, on='CLID', how='inner')
data['Pam50.Call'] = data['Pam50.Call'].astype('category')
data
```

# Analysis

## Missing data

```python
print("\nValores faltantes por columna:")
missing_values = data.isnull().sum()
print(missing_values[missing_values > 0])
```

## Assays

```python
excluded_columns = ['CLID', 'Pam50.Call']
columns_to_plot = [col for col in data.columns if col not in excluded_columns]

# Iterar sobre las columnas para graficar
for column in columns_to_plot:
    plt.figure(figsize=(8, 5))

    # Verificar si la columna es numérica o categórica
    if pd.api.types.is_numeric_dtype(data[column]):
        # Histograma para columnas numéricas
        sns.histplot(data[column], kde=True, bins=30, color='blue', edgecolor='black')
        plt.title(f"Histograma de {column}")
        plt.xlabel(column)
        plt.ylabel("Frecuencia")
    else:
        # Gráfico de barras para columnas categóricas
        sns.countplot(y=data[column], palette='viridis', order=data[column].value_counts().index)
        plt.title(f"Gráfico de barras de {column}")
        plt.xlabel("Frecuencia")
        plt.ylabel(column)

    # Mostrar el gráfico
    plt.tight_layout()
    plt.show()
```

## Relación de las varibles con Pam50.call

```python
data['Pam50_numeric'] = data['Pam50.Call'].cat.codes

excluded_columns = ['CLID', 'Pam50.Call']
columns_to_plot = [col for col in data.columns if col not in excluded_columns]

for column in columns_to_plot:
    plt.figure(figsize=(8, 5))

    if pd.api.types.is_numeric_dtype(data[column]):
        sns.regplot(x=data[column], y=data['Pam50_numeric'], scatter_kws={'alpha': 0.5})
        plt.title(f"Relación entre {column} y Pam50_numeric (Scatter)")
        plt.xlabel(column)
        plt.ylabel("Pam50_numeric")
    else:
        sns.boxplot(x=data[column], y=data['Pam50_numeric'], palette='viridis')
        plt.title(f"Relación entre {column} y Pam50_numeric (Boxplot)")
        plt.xlabel(column)
        plt.ylabel("Pam50_numeric")

    plt.tight_layout()
    plt.show()
```

### Variables y su relevancia para el modelo

#### 1. `ROR-S Group (Subtype Only)`
- **Significado:** Riesgo de recurrencia basado únicamente en el subtipo molecular.
- **Uso médico:** Es útil como referencia clínica porque categoriza pacientes en bajo, medio o alto riesgo según su subtipo.
- **Relación con la predicción directa:** Es una derivación directa del subtipo molecular (`Pam50.Call`), por lo que no debería incluirse en el modelo como variable predictora.
- **Conclusión:** **No incluir.**

#### 2. `ROR-P Group (Subtype + Proliferation)`
- **Significado:** Riesgo de recurrencia ajustado por proliferación celular.
- **Uso médico:** Combina información del subtipo molecular y de la proliferación, proporcionando una mejor evaluación del riesgo.
- **Relación con la predicción directa:** Está muy relacionado con los subtipos moleculares y la proliferación celular, lo que puede introducir redundancia en el modelo.
- **Conclusión:** **No incluir.**

#### 3. `GHI_RS_3Group` y `GHI_RS Score`
- **Significado:**  
  - `GHI_RS_3Group:` Clasifica a los pacientes en grupos de bajo, intermedio o alto riesgo según la puntuación de recurrencia del ensayo genómico Oncotype DX.  
  - `GHI_RS Score:` Puntuación continua que mide el riesgo de recurrencia.
- **Uso médico:** Evaluación pronóstica basada en la expresión de genes relacionados con el crecimiento tumoral y la proliferación.
- **Relación con la predicción directa:** Aunque está relacionado con la biología molecular del cáncer, no deriva directamente del subtipo molecular. Podría enriquecer el modelo, ya que proporciona información adicional sobre el riesgo, especialmente para los subtipos hormonales.
- **Conclusión:** **Incluir `GHI_RS Score` como variable complementaria, pero no `GHI_RS_3Group` (agrupada).**

#### 4. `Mammaprint Predict.type` y `Mammaprint Pcorr_NKI70_Good_Correlation_Nature.2002_PMID.11823860`
- **Significado:**  
  - `Mammaprint Predict.type:` Clasifica a los pacientes en bajo o alto riesgo según el ensayo genómico MammaPrint.  
  - `Mammaprint Pcorr_NKI70_Good_Correlation:` Correlación de la muestra con un perfil de buenos resultados basado en 70 genes.
- **Uso médico:** Ayuda a predecir el riesgo de recurrencia y el pronóstico a largo plazo.
- **Relación con la predicción directa:**  
  - `Mammaprint Predict.type` es una variable categórica derivada que no aporta información adicional en comparación con el score continuo.  
  - `Mammaprint Pcorr` podría enriquecer el modelo proporcionando información complementaria sobre el comportamiento del tumor.
- **Conclusión:** **Incluir `Mammaprint Pcorr`, pero no `Mammaprint Predict.type`.**

#### 5. Variables de correlación UNC
- **Significado:** Correlación de la muestra con los perfiles genómicos de los subtipos moleculares definidos por UNC (Basal, HER2-enriched, Luminal A, Luminal B, Normal-like).  
  - Variables: `UNC_Scorr_Basal_Correlation_JCO.2009_PMID.19204204`, `UNC_Scorr_Her2_Correlation_JCO.2009_PMID.19204204`, `UNC_Scorr_LumA_Correlation_JCO.2009_PMID.19204204`, `UNC_Scorr_LumB_Correlation_JCO.2009_PMID.19204204`, `UNC_Scorr_Norm_Correlation_JCO.2009_PMID.19204204`.
- **Uso médico:** Son predictores directos del subtipo molecular, ya que están diseñados para correlacionarse con los perfiles genéticos que definen cada subtipo.
- **Relación con la predicción directa:** Estas variables prácticamente predicen directamente `Pam50.Call` y no deberían incluirse en el modelo.
- **Conclusión:** **No incluir.**

#### 6. `UNC_ROR_S_Model_JCO.2009_PMID.19204204`
- **Significado:** Riesgo de recurrencia basado en el subtipo molecular según el modelo UNC.
- **Uso médico:** Es una interpretación derivada del subtipo molecular y los niveles de riesgo asociados.
- **Relación con la predicción directa:** Es una variable redundante que deriva directamente del subtipo molecular y no debería incluirse.
- **Conclusión:** **No incluir.**

#### 7. `UNC_Proliferation_11_Mean_JCO.2009_PMID.19204204`
- **Significado:** Promedio de expresión de 11 genes relacionados con la proliferación celular.
- **Uso médico:** La proliferación celular es un marcador importante de agresividad tumoral. Es independiente del subtipo molecular, aunque tiene cierta correlación con subtipos como Basal-like o HER2-enriched.
- **Relación con la predicción directa:** No predice directamente el subtipo molecular, pero aporta información adicional sobre la biología del tumor.
- **Conclusión:** **Incluir como variable complementaria.**

#### 8. `ROR-P (Subtype + Proliferation)`
- **Significado:** Similar a `ROR-P Group`, ajusta el riesgo de recurrencia por proliferación celular.
- **Uso médico:** Combina información del subtipo molecular y de la proliferación.
- **Relación con la predicción directa:** Está estrechamente relacionado con el subtipo molecular y puede introducir redundancia.
- **Conclusión:** **No incluir.**
