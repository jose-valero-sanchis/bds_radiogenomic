# Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path = "../../data/"
```

# Data

```python
df_assays = pd.read_excel(data_path + "multigenic_assays.xlsx")
df_assays = df_assays.drop(columns=df_assays.columns[1])


target_class = pd.read_csv(data_path + 'target_class.csv')
```

```python
data = pd.merge(df_assays, target_class, on='CLID', how='inner')
data['Pam50.Call'] = data['Pam50.Call'].astype('category')
data
```

# Analysis

## Missing data

```python
print("\nValores faltantes por columna:")
missing_values = data.isnull().sum()
print(missing_values[missing_values > 0])
```

## Assays

```python
excluded_columns = ['CLID', 'Pam50.Call']
columns_to_plot = [col for col in data.columns if col not in excluded_columns]

# Iterar sobre las columnas para graficar
for column in columns_to_plot:
    plt.figure(figsize=(8, 5))

    # Verificar si la columna es numÃ©rica o categÃ³rica
    if pd.api.types.is_numeric_dtype(data[column]):
        # Histograma para columnas numÃ©ricas
        sns.histplot(data[column], kde=True, bins=30, color='blue', edgecolor='black')
        plt.title(f"Histograma de {column}")
        plt.xlabel(column)
        plt.ylabel("Frecuencia")
    else:
        # GrÃ¡fico de barras para columnas categÃ³ricas
        sns.countplot(y=data[column], palette='viridis', order=data[column].value_counts().index)
        plt.title(f"GrÃ¡fico de barras de {column}")
        plt.xlabel("Frecuencia")
        plt.ylabel(column)

    # Mostrar el grÃ¡fico
    plt.tight_layout()
    plt.show()
```

## RelaciÃ³n de las varibles con Pam50.call

```python
data['Pam50_numeric'] = data['Pam50.Call'].cat.codes

excluded_columns = ['CLID', 'Pam50.Call']
columns_to_plot = [col for col in data.columns if col not in excluded_columns]

for column in columns_to_plot:
    plt.figure(figsize=(8, 5))

    if pd.api.types.is_numeric_dtype(data[column]):
        sns.regplot(x=data[column], y=data['Pam50_numeric'], scatter_kws={'alpha': 0.5})
        plt.title(f"RelaciÃ³n entre {column} y Pam50_numeric (Scatter)")
        plt.xlabel(column)
        plt.ylabel("Pam50_numeric")
    else:
        sns.boxplot(x=data[column], y=data['Pam50_numeric'], palette='viridis')
        plt.title(f"RelaciÃ³n entre {column} y Pam50_numeric (Boxplot)")
        plt.xlabel(column)
        plt.ylabel("Pam50_numeric")

    plt.tight_layout()
    plt.show()
```

### Variables y su relevancia para el modelo

#### 1. `ROR-S Group (Subtype Only)`
- **Significado:** Riesgo de recurrencia basado Ãºnicamente en el subtipo molecular.
- **Uso mÃ©dico:** Es Ãºtil como referencia clÃ­nica porque categoriza pacientes en bajo, medio o alto riesgo segÃºn su subtipo.
- **RelaciÃ³n con la predicciÃ³n directa:** Es una derivaciÃ³n directa del subtipo molecular (`Pam50.Call`), por lo que no deberÃ­a incluirse en el modelo como variable predictora.
- **ConclusiÃ³n:** **No incluir.**

#### 2. `ROR-P Group (Subtype + Proliferation)`
- **Significado:** Riesgo de recurrencia ajustado por proliferaciÃ³n celular.
- **Uso mÃ©dico:** Combina informaciÃ³n del subtipo molecular y de la proliferaciÃ³n, proporcionando una mejor evaluaciÃ³n del riesgo.
- **RelaciÃ³n con la predicciÃ³n directa:** EstÃ¡ muy relacionado con los subtipos moleculares y la proliferaciÃ³n celular, lo que puede introducir redundancia en el modelo.
- **ConclusiÃ³n:** **No incluir.**

#### 3. `GHI_RS_3Group` y `GHI_RS Score`
- **Significado:**  
  - `GHI_RS_3Group:` Clasifica a los pacientes en grupos de bajo, intermedio o alto riesgo segÃºn la puntuaciÃ³n de recurrencia del ensayo genÃ³mico Oncotype DX.  
  - `GHI_RS Score:` PuntuaciÃ³n continua que mide el riesgo de recurrencia.
- **Uso mÃ©dico:** EvaluaciÃ³n pronÃ³stica basada en la expresiÃ³n de genes relacionados con el crecimiento tumoral y la proliferaciÃ³n.
- **RelaciÃ³n con la predicciÃ³n directa:** Aunque estÃ¡ relacionado con la biologÃ­a molecular del cÃ¡ncer, no deriva directamente del subtipo molecular. PodrÃ­a enriquecer el modelo, ya que proporciona informaciÃ³n adicional sobre el riesgo, especialmente para los subtipos hormonales.
- **ConclusiÃ³n:** **Incluir `GHI_RS Score` como variable complementaria, pero no `GHI_RS_3Group` (agrupada).**

#### 4. `Mammaprint Predict.type` y `Mammaprint Pcorr_NKI70_Good_Correlation_Nature.2002_PMID.11823860`
- **Significado:**  
  - `Mammaprint Predict.type:` Clasifica a los pacientes en bajo o alto riesgo segÃºn el ensayo genÃ³mico MammaPrint.  
  - `Mammaprint Pcorr_NKI70_Good_Correlation:` CorrelaciÃ³n de la muestra con un perfil de buenos resultados basado en 70 genes.
- **Uso mÃ©dico:** Ayuda a predecir el riesgo de recurrencia y el pronÃ³stico a largo plazo.
- **RelaciÃ³n con la predicciÃ³n directa:**  
  - `Mammaprint Predict.type` es una variable categÃ³rica derivada que no aporta informaciÃ³n adicional en comparaciÃ³n con el score continuo.  
  - `Mammaprint Pcorr` podrÃ­a enriquecer el modelo proporcionando informaciÃ³n complementaria sobre el comportamiento del tumor.
- **ConclusiÃ³n:** **Incluir `Mammaprint Pcorr`, pero no `Mammaprint Predict.type`.**

#### 5. Variables de correlaciÃ³n UNC
- **Significado:** CorrelaciÃ³n de la muestra con los perfiles genÃ³micos de los subtipos moleculares definidos por UNC (Basal, HER2-enriched, Luminal A, Luminal B, Normal-like).  
  - Variables: `UNC_Scorr_Basal_Correlation_JCO.2009_PMID.19204204`, `UNC_Scorr_Her2_Correlation_JCO.2009_PMID.19204204`, `UNC_Scorr_LumA_Correlation_JCO.2009_PMID.19204204`, `UNC_Scorr_LumB_Correlation_JCO.2009_PMID.19204204`, `UNC_Scorr_Norm_Correlation_JCO.2009_PMID.19204204`.
- **Uso mÃ©dico:** Son predictores directos del subtipo molecular, ya que estÃ¡n diseÃ±ados para correlacionarse con los perfiles genÃ©ticos que definen cada subtipo.
- **RelaciÃ³n con la predicciÃ³n directa:** Estas variables prÃ¡cticamente predicen directamente `Pam50.Call` y no deberÃ­an incluirse en el modelo.
- **ConclusiÃ³n:** **No incluir.**

#### 6. `UNC_ROR_S_Model_JCO.2009_PMID.19204204`
- **Significado:** Riesgo de recurrencia basado en el subtipo molecular segÃºn el modelo UNC.
- **Uso mÃ©dico:** Es una interpretaciÃ³n derivada del subtipo molecular y los niveles de riesgo asociados.
- **RelaciÃ³n con la predicciÃ³n directa:** Es una variable redundante que deriva directamente del subtipo molecular y no deberÃ­a incluirse.
- **ConclusiÃ³n:** **No incluir.**

#### 7. `UNC_Proliferation_11_Mean_JCO.2009_PMID.19204204`
- **Significado:** Promedio de expresiÃ³n de 11 genes relacionados con la proliferaciÃ³n celular.
- **Uso mÃ©dico:** La proliferaciÃ³n celular es un marcador importante de agresividad tumoral. Es independiente del subtipo molecular, aunque tiene cierta correlaciÃ³n con subtipos como Basal-like o HER2-enriched.
- **RelaciÃ³n con la predicciÃ³n directa:** No predice directamente el subtipo molecular, pero aporta informaciÃ³n adicional sobre la biologÃ­a del tumor.
- **ConclusiÃ³n:** **Incluir como variable complementaria.**

#### 8. `ROR-P (Subtype + Proliferation)`
- **Significado:** Similar a `ROR-P Group`, ajusta el riesgo de recurrencia por proliferaciÃ³n celular.
- **Uso mÃ©dico:** Combina informaciÃ³n del subtipo molecular y de la proliferaciÃ³n.
- **RelaciÃ³n con la predicciÃ³n directa:** EstÃ¡ estrechamente relacionado con el subtipo molecular y puede introducir redundancia.
- **ConclusiÃ³n:** **No incluir.**
