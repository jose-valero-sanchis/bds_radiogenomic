```python
import os
import SimpleITK as sitk
import pydicom
import matplotlib.pyplot as plt
```

```python
def load_all_series_in_folder(patient_folder):
    images = {}
    reader = sitk.ImageSeriesReader()
    
    for root, dirs, files in os.walk(patient_folder):
        series_ids = reader.GetGDCMSeriesIDs(root)
        for s_id in series_ids:
            file_names = reader.GetGDCMSeriesFileNames(root, s_id)
            if not file_names:
                continue
            reader.SetFileNames(file_names)
            sitk_image = reader.Execute()
            images[s_id] = (sitk_image, file_names)
    return images
```

```python
base_dir = "../../data/images/manifest-AmUKkZHx1564984923148296294/TCGA-BRCA/"
patient_folder = os.path.join(base_dir, "TCGA-AO-A0JI", "11-06-2001-NA-NA-80329")

all_series = load_all_series_in_folder(patient_folder)

tags_of_interest = [
    "SeriesDescription", 
    "Modality",
    "ProtocolName",
    "SeriesNumber",
    "StudyDescription",
    "Manufacturer",
    "InstitutionName",
    "PatientID",
    "PatientName",
    "PatientSex",
    "PatientAge",
    "PatientBirthDate",
    "StudyDate",
    "StudyTime",
    "AccessionNumber",
    "PixelSpacing",
    "SpacingBetweenSlices",
    "SliceThickness",
    "RepetitionTime",
    "EchoTime",
    "FlipAngle",
]

for s_id, (sitk_image, file_names) in all_series.items():
    size = sitk_image.GetSize()       # (width, height, num_slices) en ITK (ojo con el orden)
    spacing = sitk_image.GetSpacing() # (espaciado_x, espaciado_y, espaciado_z)
    origin = sitk_image.GetOrigin()   # punto de origen
    direction = sitk_image.GetDirection() # matriz de direcci√≥n
    print("="*80)
    print(f"Serie UID: {s_id}")
    print(f"  Tama√±o (GetSize): {size}")
    print(f"  Spacing (GetSpacing): {spacing}")
    print(f"  Origin (GetOrigin): {origin}")
    print(f"  Direction (GetDirection): {direction}")

    # Leemos metadatos DICOM de (al menos) un archivo para ver la descripci√≥n de la serie
    first_dicom_file = file_names[0]  # Tomamos cualquier archivo de la serie
    ds = pydicom.dcmread(first_dicom_file, stop_before_pixels=True)  # para no leer la imagen en s√≠
    print("  -- Metadatos DICOM relevantes --")
    for tag in tags_of_interest:
        value = getattr(ds, tag, "N/A")
        print(f"     {tag}: {value}")
    
    # OPCIONAL: Si quieres visualizar un corte de la serie
    # (a modo de ejemplo, mostramos la rebanada central)
    arr = sitk.GetArrayFromImage(sitk_image)  # shape = (num_slices, alto, ancho)
    mid_slice = arr.shape[0] // 2
    plt.imshow(arr[mid_slice, :, :], cmap='gray')
    plt.title(f"Serie: {s_id}\n{getattr(ds, 'SeriesDescription', 'N/A')}")
    plt.show()
```
